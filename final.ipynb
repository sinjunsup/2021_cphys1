## 1. Random Walks

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import collections
```

## (a)

```python
def where(t): # 시간에 따른 x위치
    x=0
    for h in range(t):
        if random.random()<1/2:
            x=x -1 
        else:
            x=x + 1
    return x
```

## (b)

```python
def aver(n,t): # 시간 t에 따른 n번 시도
    times = np.zeros(n)
    for h in range(n):
        times[h]=where(t)
    return times
```

```python
def bolt(n,t,k): # 갯수세는함수
    ct=aver(n,t)
    y=np.count_nonzero(ct == k)
    return y
```

```python
k=np.arange(-15,15,1)
plt.plot(k,[bolt(100,0,i) for i in k],label='t=0') #100개 시간0
plt.plot(k,[bolt(100,10,i) for i in k],label='t=10')#100개 시간10
plt.plot(k,[bolt(100,50,i) for i in k],label='t=50')#100개 시간50
plt.plot(k,[bolt(100,100,i) for i in k],label='t=100')#100개 시간100
plt.legend()

plt.xlabel("x")
plt.ylabel('n')
plt.show()
```

## 2. Bifurcation

```python
import numpy as np
import sympy
from sympy import *
```

## (a) 이 방정식을 다음의 조건에서 풀고 궤적을 그려라. (미분방정식)

```python
t = symbols('t')
x = Function('x')
a=dsolve(Eq(x(t).diff(t),-1*x(t)-x(t)*x(t)),ics= {x(0):2})# 초기조건 x(0) = 2, r=-1
b=dsolve(Eq(x(t).diff(t),-1*x(t)-x(t)*x(t)),ics= {x(0):-2})# 초기조건 x(0) = -2, r=-1
c=dsolve(Eq(x(t).diff(t),1*x(t)-x(t)*x(t)),ics= {x(0):2})# 초기조건 x(0) = 2, r=1
d=dsolve(Eq(x(t).diff(t),1*x(t)-x(t)*x(t)),ics= {x(0):-2})# 초기조건 x(0) = -2, r=1
```

```python
sympy.simplify(a),sympy.simplify(b),sympy.simplify(c),sympy.simplify(d)
```

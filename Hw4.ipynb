## 1. Error Function

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
def exp(x):
    return (2/(np.pi**0.5))*np.exp(-x**2)
```

```python
def er(f,x,dt):
    t=0
    s=0
    while(t<abs(x)):
        s= s+0.5*dt*(f(t+dt)+f(t))
        t = t+ dt
    return x/abs(x)*s
```

```python
k=np.arange(-2,2,0.01)
plt.plot(k,[er(exp,i,0.01) for i in k])
plt.ylim(-1,1)
```

```python
def er2(x,p):
    n=0
    t=0
    while(n<p):
        r=n
        fac=1
        for k in range(1,r+1):
            fac =fac*k
        t= t+((-1)**n)*(x**(2*n+1))/(fac*(2*n+1))
        
        n=n+1
    return (2/np.pi**0.5)*t
```

```python
k=np.arange(-2,2,0.01)
plt.plot(k,[er2(i,20) for i in k])
plt.ylim(-1,1)
```

## (Integration) Trigonometric Function

```python
def s(x):
    return np.sin(x)
def c(x):
    return np.cos(x)
```

```python
def int(g,a,b,dx):
    x=a
    s=0
    while(x<b):
        s= s+0.5 *dx*(g(x+dx)+g(x))
        x = x+dx
    return s
```

```python
int(s,0,4,0.1)
```

```python
integrate.quad(s,0,4)
```

```python
x = np.arange(0,5,0.1)
plt.plot(x,[int(s,0,i,0.01) for i in x], label= 'sin')
plt.plot(x,[int(c,0,i,0.01) for i in x], label='cos')
plt.legend()
```

```python
x = np.arange(0,5,0.1)
plt.plot(x,[integrate.quad(s,0,i) for i in x],label='sin')
plt.plot(x,[integrate.quad(c,0,i) for i in x],label='cos')
plt.legend()
```

```python
x = np.arange(0,5,0.1)
plt.plot(x,[int(s,0,i,0.01) for i in x], label= 'sin1')
plt.plot(x,[int(c,0,i,0.01) for i in x], label='cos1')
plt.plot(x,[integrate.quad(s,0,i) for i in x],label='sin2')
plt.plot(x,[integrate.quad(c,0,i) for i in x],label='cos2')
plt.legend()
```

## 3. (Differentiation) Trigonometric Function

```python
def front(f,x,dx):
    return(f(x+dx)-f(x))/dx
def back(f,x,dx):
    return (f(x)-f(x-dx))/dx
def mid(f,x,dx):
    return(f(x+dx)/2 - f(x-dx)/2)/dx
```

```python
x=np.arange(0,5,0.1)
plt.plot(x,front(s,x,0.1),label='for')
plt.plot(x,back(s,x,0.1),label='back')
plt.plot(x,mid(s,x,0.1),label='cen')
plt.legend()
```

```python
x=np.arange(0,5,0.1)
plt.plot(x,front(c,x,0.1),label='for')
plt.plot(x,back(c,x,0.1),label='back')
plt.plot(x,mid(c,x,0.1),label='cen')
plt.legend()
```

```python
from scipy.misc import derivative
plt.plot(x,derivative(s,x,dx=0.01,n=1),label='desin')
plt.plot(x,derivative(c,x,dx=0.01,n=1),label='decos')
plt.legend()
```

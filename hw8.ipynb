## 1. Hypersphere

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import collections
```

```python
def hypershpere(d,n): # 차원에 따른 Cn값 
    states = np.ones(d)
    count=0
    for i in range(n):
        for i in range(d):
            r = random.random()
            states[i] = r*r
            
        if sum(states) < 1:
            count+=1
    return (2**d)*count/n
```

```python
x= [10,100,1000,10000,100000,1000000] # 3차원
p = [hypershpere(3,i) for i in x]
plt.plot(x,p)
```

## Metropolis Algorithm

## 1. 총 에너지

```python
import random
import numpy as np
import matplotlib.pyplot as plt 

def aveE(b,n): # 온도에 따른 메트로폴리즘 알리고리즘
    states = np.ones(n)
    for t in range(n*100):
        i=random.randint(0,n-1)
        DE=0
        if random.random()<1/2:
            DE = -1
        else:
            DE = 1
            
        if DE < 0:  
            states[i]=states[i] -1 
        elif random.random()<np.exp(-b*DE):
            states[i]=states[i] + 1
        if states[i]==0:
            states[i]=1
    return states
```

```python
temp= np.arange(1,20,1)
y = [sum(aveE(1/b,1000)**2) for b in temp]
plt.plot(temp,y,'o')
```

```python
temp= np.arange(1,20,1)
y = [sum(aveE(b,1000)**2) for b in temp]
plt.plot(temp,y,'o')
```

## 2. 에너지에 따른 입자의 분포 측정

```python
def bolt(b,n): #   에너지크기에 따른 입자 분포세는 함수
    ct=aveE(b,n)
    nb = np.ones(11)
    for t in range(1,10):
        nb[t]=np.count_nonzero(ct == t)
    return nb       
```

```python
plt.plot(bolt(1,1000),label='b=1') # b가 1일때
plt.plot(bolt(0.5,1000),label='b=0.5')# b가 0.5일때
plt.plot(bolt(4,1000),label='b=4')# b가 4일때
plt.legend()

plt.xlabel("E")
plt.ylabel('number')
plt.show()
```

## 1. Pendulum

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.001 #a 는 세타 , w는 각속도
t = np.arange(0,100,dt)
a , w = np.array([]), np.array([])
a0, w0 = 1, 0 # 초기위치, 초기위상

for i in t:
    a0 = a0 + dt*w0
    w0 = w0 - dt*1*np.sin(a0) # g/L 은 1로 가정
    a=np.append(a,a0)
    w=np.append(w,w0)
```

```python
plt.plot(t,a)
```

## 4. Forced Pendulum

```python
dt=0.001 #a 는 세타 , w는 각속도
A, v = 1,1 #A는 외부힘 v는 외부진동수
g,l= 1,1  #g 중력가속도  ,l 진자의 길이
t = np.arange(0,100,dt)
a , w = np.array([]), np.array([])
a0, w0 = 1, 0 # 초기위치1,초기위상
b= 0.1 

for i in t:
    a0 = a0 + dt*w0
    w0 = w0 - dt*w0*b - dt*(g/l)*np.sin(a0) + A*np.cos(v*i)
    a=np.append(a,a0)
    w=np.append(w,w0)
plt.plot(t,a)
```
